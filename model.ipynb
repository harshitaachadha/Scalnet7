```python
import tensorflow as tf
import os
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import numpy as np

def load_images(path):
    image = tf.io.read_file(path)
    image = tf.io.decode_jpeg(image)
    image = image[:,:,:3]
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, (224,224))
    image = image/255.0
    return image
```

```python
from os import listdir
import numpy as np
from skimage import data, io, filters
filebefore = '/content/drive/MyDrive/Major_Project/python_code_major/scalograms/scal_images/'
elec_1_list = [f for f in listdir('/content/drive/MyDrive/Major_Project/python_code_major/scalograms/scal_images') ]
Y = [] 
X = []
total=len(elec_1_list)
i=1
n=0
m=0
finlist = []
for item in elec_1_list:
    if 'schiz'in item:
      if(n<125):
        finlist.append(item)
        n=n+1
    else:
      if(m<125):
        finlist.append(item)
        m=m+1
print(finlist)
for item in finlist:
    if 'schiz'in item:
      Y.append(1)
    else:
      Y.append(0)
    sample_path = filebefore+item
    x = load_images(sample_path)
    X.append(x)
    i+=1  
```

```python
import numpy
X_arr = numpy.array(X)
Y_arr = numpy.array(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_arr, Y_arr, test_size=0.20, random_state=5)
```

```python
X_train.shape
```

```python
from keras.layers import *
from keras.models import Sequential
from os import listdir
model = Sequential()

model.add(Conv2D(filters=16,kernel_size=2,padding="same",activation="relu",input_shape=(224,224,3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation ="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Flatten())
model.add(Dense(500,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
#training the model
model.fit(X_train,y_train,epochs=20,validation_split=0.1,batch_size=16, )
```

```python
model.evaluate(X_test,y_test)
```

```python
import sys
from os import listdir
from PIL import Image
schiz_list = []
for i in range(0,84):
  num = i+1
  part1 = 'per'+str(num)+'_'
  if i>44:
    part2 = 'norm_'
  else:
    part2 = 'schiz_'
  part3 = '1_'
  for j in range(0,14):
    w = j+1
    part4=str(w)+'.png'
    total = part1+part2+part3+part4
    schiz_list.append(total)
print(schiz_list)
final = []
for item in schiz_list:
  final.append('/content/drive/MyDrive/python_code_major/electrode_1/'+str(item))

first = 0
last = 14
for i in range(0,84):
  images = [Image.open(x) for x in final[first:last]]
  widths, heights = zip(*(i.size for i in images))

  total_width = sum(widths)
  max_height = max(heights)

  new_im = Image.new('RGB', (total_width, max_height))

  x_offset = 0
  for im in images:
    new_im.paste(im, (x_offset,0))
    x_offset += im.size[0]
  base = '/content/drive/MyDrive/python_code_major/concat_image/electrode_1/'
  added = 'per_'+str(i+1)+'.jpg'
  totals = base+added
  new_im.save(totals)
  first = last
  last = first+14
```
